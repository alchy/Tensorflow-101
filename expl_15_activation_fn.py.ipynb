# Various activiation functions for Perceptron

Activation functions bring non-linear properties to our perceptron. Bellow I present three most popular at the date.

```python
# name:   expl_08_activation_fn.py
# v0.1/jn

import matplotlib.pyplot as plt
import numpy as np

# sigmoid or logistic
#
#        logistic sigmoid can cause a neural network to get 
#        stuck during training. this is due in part to the fact 
#        that if a strongly-negative input is provided to the 
#        logistic sigmoid, it outputs values very near to zero
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Tanh — Hyperbolic tangent
def tanh(x):
    return np.tanh(x)

# ReLu, rectified linear units
#
#        almost all deep learning models use ReLu nowadays, but
#        the problem of ReLu is that during the training some 
#        ReLu perceptrons die - if their w is very close to zero
#        * leaking ReLu is the way and TF implements lru
def relu(x):
    return (abs(x) + x) / 2


plt.style.use('ggplot')

x = np.arange(-2.5, 2.5, .25)

plt.plot(x, sigmoid(x), label='sigmoid')
plt.plot(x, tanh(x), label='tanh')
plt.plot(x, relu(x), label='relu')

plt.legend(loc='best')
plt.show()
```
