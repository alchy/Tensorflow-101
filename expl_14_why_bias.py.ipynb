# Role of Bias

Bias is similar to the constant b of a linear function y = ax + b. Neuron needs linear function to do simple classification. If b == 0 then linear function intersects zero coordinates [0, 0].

![alt text](pictures/220px-Linear_Function_Graph.svg.png "linear function (https://en.wikipedia.org/wiki/Linear_function)")

To to make it easier, let's say, we have a simple neuron with a single input. The math inside our simple neuron is just a linear function followed by an activation function hooked at the output. In our first example our linear function has no b (bias).

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('ggplot')

def sigmoid_fn(inputs):
    ''' compute sigmoid fn: 1 / (1 + math.e ** -x) '''
    return([1 / float(1 + np.exp(- x)) for x in inputs])

x = np.arange(-20., 20., 0.2)
weight = 1.

y = sigmoid_fn(weight * x)
plt.plot(x, y, label='w/o bias')
    
plt.legend(loc='best')
plt.show()
```

As you may see, the activation function is pretty centered. Adding bias allows you to shift the activation function to the left or right. This boosts perceptron's classification performance. The example below shows output of the activation function for various biases.

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('ggplot')

def sigmoid_fn(inputs):
    ''' compute sigmoid fn: 1 / (1 + math.e ** -x) '''
    return([1 / float(1 + np.exp(- x)) for x in inputs])

x = np.arange(-20., 20., 0.2)
w = 1.
b = [-10., .0, 10.] 

for bias in b:
    y = sigmoid_fn((w * x) + bias)
    plt.plot(x, y, label='w/o bias %f' % bias)
    
plt.legend(loc='best')
plt.show()
```

The bias is mandatory for neuron to perform an effective classification. In NN, bias can be a trainable Variable or a constant having it's own weight.
